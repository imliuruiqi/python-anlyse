```python
import json
import pandas as pd
import numpy as np
```

```python
#导入文件并读取json至列表data
file = 'IPC.json'
with open(file,'r',encoding='utf-8') as f:
    data = json.load(f)
data
```

```python
codeli = []#初始化列表
for dic in data:
    chili = dic['classifyNumRoot']['children']
    for child in chili:
        code = {}
        code['ID'] = child['id']
        code['部']= child['utcn'].split(';')[0][3:]
        code['类'] = child['utcn'].split(';')[1][4:]
        code['department']= child['uten'].split(';')[0]
        code['class'] = child['uten'].split(';')[1][4:]
        codeli.append(code)
codeli
```

```python
IPCdf = pd.DataFrame(codeli)
IPCdf
```

```python
IPCdf.to_excel('IPC.xls',sheet_name='sheet1')#导出
```
