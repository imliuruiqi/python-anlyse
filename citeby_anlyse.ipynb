```python
# 对引用专利进行分析
# 数据已经过 cited_data脚本 统一处理

import pandas as pd
import numpy as np
from chinese_province_city_area_mapper import drawers

# 导入文件并转化为pandas DataFrame对象
d = pd.read_excel('citebywithcountry.xls',sheet_name='sheet1')
do = pd.read_excel('citebywithcountry.xls',sheet_name='original')
dc = pd.read_excel('citebywithcountry.xls',sheet_name='citeby')

# 打印引用原290条专利的专利，共62条记录，即总引用次数达62次。
dc
```

```python
#分析这62条记录的特征

# 第一申请人地址（国家）分布
countrycount = dc['country'].value_counts()
countrycount_o = do['country'].value_counts()

# 同族成员数分布
familycount = dc['INPADOC同族成员数量'].value_counts()
familycount_o = do['INPADOC同族成员数量'].value_counts()

# 申请年分布
yearcount = dc['申请年'].value_counts()
yearcount.sort_index(inplace=True)
yearcount_o = do['申请年'].value_counts()
yearcount_o.sort_index(inplace=True)

# IPC主分类号（大类）分布
classcount = dc['IPC主分类号(大类)'].value_counts()
classcount_o = do['IPC主分类号(大类)'].value_counts()

# IPC主分类号(部)分布
partcount = dc['IPC主分类号(部)'].value_counts()
partcount_o = do['IPC主分类号(部)'].value_counts() 
```

```python
# 画图分析

from pyecharts import Bar
from pyecharts import Bar
countrybar = Bar('引用研究对象的专利的国家分布')
countrybar.add('', countrycount.index, countrycount.values, 
               is_label_show=True, xaxis_interval=0, xaxis_rotate=-80, mark_line=['average'])
countrybar
```

```python
familybar = Bar('INPADOC同族成员数量分布')
familybar.add('', familycount.index, familycount.values, 
               is_label_show=True, xaxis_interval=0, xaxis_rotate=-80, mark_line=['average'])
familybar
```

```python
year_obar = Bar('原研究对象的申请年份分布')
year_obar.add('原研究对象', yearcount_o.index, yearcount_o.values, 
               is_label_show=True, xaxis_interval=0, xaxis_rotate=-80, mark_line=['average'])
year_obar
```

```python
yearbar = Bar('引用研究对象的专利的申请年份分布')
yearbar.add('引用研究对象的专利', yearcount.index, yearcount.values, 
               is_label_show=True, xaxis_interval=0, xaxis_rotate=-80, mark_line=['average'])
#yearbar.add('原研究对象', yearcount_o.index, yearcount_o.values, 
#               is_label_show=True, xaxis_interval=0, xaxis_rotate=-80, mark_line=['average'])
yearbar
```

```python
classbar = Bar('IPC主分类号(大类)数量分布')
classbar.add('', classcount.index, classcount.values, 
               is_label_show=True, xaxis_interval=0, xaxis_rotate=-80, mark_line=['average'])
classbar
```
