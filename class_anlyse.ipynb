```python
import pandas as pd
import numpy as np

# 引用分析

d = pd.read_excel('citewithcountry.xls',sheet_name='sheet1')
do = pd.read_excel('citewithcountry.xls',sheet_name='original')
dc = pd.read_excel('citewithcountry.xls',sheet_name='citeby')
```

```python
classcount = dc['IPC主分类号(大类)'].value_counts()
classli = list(classcount.index)
classli.remove('-')
```

```python
classdict = {}
for classcode in classli:
    dolist = list(do['公开(公告)号'][do['IPC主分类号(大类)']== classcode])
    classdf = dc[dc['公开(公告)号'].isin(dolist)]
    classdf['citeclass'] = classcode
    classdict[classcode] = classdf
```

```python
# 筛选记录数大于10的作为Topclass分析
topclassdict = {key: value for key,value in classdict.items() if len(value)>=10}
for key,value in topclassdict.items():print(key,len(value))
```

```python
classd = topclassdict['C08']

# IPC主分类号（大类）分布
classcount_cl = d['IPC主分类号(大类)'].value_counts()
# 第一申请人地址（国家）分布
countrycount_cl = dc['country'].value_counts()
```

```python
# 画图分析

from pyecharts import Bar
country_clbar = Bar('C08类目引用的专利的国家分布')
country_clbar.add('', countrycount_cl.index, countrycount_cl.values, 
               is_label_show=True, xaxis_interval=0, xaxis_rotate=-80, mark_line=['average'])
country_clbar
```

```python
class_clbar = Bar('C08类目引用的专利的IPC主分类号(大类)分布')
class_clbar.add('', classcount_cl.index, classcount_cl.values, 
               is_label_show=True, xaxis_interval=0, xaxis_rotate=-80, mark_line=['average'])
class_clbar
```
